**Performming K Means Clustering**

```python
train_images=loaded_images
```

```python
clustered_img = []
cluster_dict = {}
```

```python
for j in range(0,348):
  print("Image "+str(j)+" Started")
  k_mean_test_img=train_images[j]
  # Convert the image to the Lab color space
  lab_image = rgb2lab(k_mean_test_img)
  # Reshape the image into a 2D array
  X = lab_image.reshape((lab_image.shape[0] * lab_image.shape[1], 3))

  # Perform K-means clustering with 10 clusters
  kmeans = KMeans(n_clusters=10)
  kmeans.fit(X)

  # Get the cluster labels for each pixel
  labels = kmeans.labels_

  # Reshape the labels into a 2D array
  labels_2d = labels.reshape((lab_image.shape[0], lab_image.shape[1]))

  # Create a new image with the same dimensions as the original image
  new_image = np.zeros_like(k_mean_test_img)

  # Assign a different color to each cluster
  for i in range(10):
    new_image[labels_2d == i] = [255, 0, 0] if i == 0 else [0, 255, 0] if i == 1 else [0, 0, 255] if i == 2 else [255, 255, 0] if i == 3 else [0, 255, 255] if i == 4 else [255, 0, 255] if i == 5 else [128, 0, 0] if i == 6 else [0, 128, 0] if i == 7 else [0, 0, 128] if i == 8 else [128, 128, 0] if i == 9 else [0, 128, 128]
  # Find the area of each cluster
  areas = []
  for i in range(10):
    area = np.count_nonzero(labels_2d == i)
    areas.append(area)
  # Find the average RGB value for each cluster
  cluster_avg_rgb = []
  for i in range(10):
    cluster_pixels = k_mean_test_img[labels_2d == i]
    avg_r = np.mean(cluster_pixels[:, 0])
    avg_g = np.mean(cluster_pixels[:, 1])
    avg_b = np.mean(cluster_pixels[:, 2])
    cluster_avg_rgb.append([avg_r, avg_g, avg_b])
  # Create a new image with the same dimensions as the original image
  segmented_image = np.zeros_like(k_mean_test_img)

  # Assign each pixel of respective clusters with AVG RGB value of each cluster
  for i in range(10):
    segmented_image[labels_2d == i] = cluster_avg_rgb[i]
  clustered_img.append(segmented_image)
  cluster_dict_len=len(cluster_dict)
  for i in range(10):
    cluster_dict[i+cluster_dict_len] = {
      "avg_rgb": cluster_avg_rgb[i],
      "area": areas[i]
    }
  print("Image "+str(j)+" Ended")
```

Image cropping function definition


```python
class CropLayer(object):
    def __init__(self, params, blobs):
        # initialize our starting and ending (x, y)-coordinates of the crop
        self.startX = 0
        self.startY = 0
        self.endX = 0
        self.endY = 0

    def getMemoryShapes(self, inputs):
        # the crop layer will receive two inputs — we need to crop
        # the first input blob to match the shape of the second one,
        # keeping the batch size and number of channels
        (inputShape, targetShape) = (inputs[0], inputs[1])
        (batchSize, numChannels) = (inputShape[0], inputShape[1])
        (H, W) = (targetShape[2], targetShape[3])

        # compute the starting and ending crop coordinates
        self.startX = int((inputShape[3] - targetShape[3]) / 2)
        self.startY = int((inputShape[2] - targetShape[2]) / 2)
        self.endX = self.startX + W
        self.endY = self.startY + H

        # return the shape of the volume (we'll perform the actual crop during the forward pass)
        return [[batchSize, numChannels, H, W]]

    def forward(self, inputs):
        # use the derived (x, y)-coordinates to perform the crop
        return [inputs[0][:, :, self.startY:self.endY, self.startX:self.endX]]
```
